```python
#Name of folder containing relevant data:
folderName = 'data/taylor814Dec3/'

#Must have scikits audiolab installed, this can be a bit tricky on osx
from scikits.audiolab import Format, Sndfile, play, wavread, wavwrite
import sys, time

#File names to import:
fileNames = ['C451-05', 'DFLY-03', 'DI-06', 'KM84-06']

sampleRates = np.zeros(len(fileNames))
allAudio = []

for i, fileName in enumerate(fileNames):
    data, fs, enc = wavread(folderName + '/' + fileName + '.wav')
    sampleRates[i] = fs
    allAudio.append(data)
    
#Let's check to make sure sample rates are consistent:
for i in range(len(fileNames)):
    if sampleRates[i] != np.median(sampleRates):
        print "SHIT, SAMPLE RATES DON'T MATCH"
    else:
        Fs = np.median(sampleRates)
        
Fs = Fs.astype('int')
```

##Sample portion of recoriding
- And asign names to each track

```python
#In seconds:
startingPoint = 5
endingPoint = 30

microphoneOneAudio = allAudio[0][startingPoint*Fs:endingPoint*Fs]
microphoneTwoAudio = allAudio[1][startingPoint*Fs:endingPoint*Fs]
piezoAudio = allAudio[2][startingPoint*Fs:endingPoint*Fs]
microphoneThreeAudio = allAudio[3][startingPoint*Fs:endingPoint*Fs]
```

##Normalize

```python
#Normalize to +/- 1. 
microphoneOneAudio = microphoneOneAudio/max(abs(microphoneOneAudio))
microphoneTwoAudio = microphoneTwoAudio/max(abs(microphoneTwoAudio))
microphoneThreeAudio = microphoneThreeAudio/max(abs(microphoneThreeAudio))
piezoAudio = piezoAudio/max(abs(piezoAudio))
```

#Time alignment
- We expect a delay between piezo and microphone signals roughly corresponding to the time required for the sound to reach the microphone
- At a sampling rate of 48Khz, and taking the speed of sound to be 1130 ft/second, we expect a delay of: 

$$
\frac{48000 samples}{second}
\frac{1 second}{1130 ft} = 
42.48 \frac{samples}{foot}
$$

There are clearly a number of ways to numerically approach audio alignment. We'll try two here: 
1. Maximizing correlation of time-aligned signals
2. Minimizing error of traiend OLS model on time-aligned signals

These analyses can be pretty slow, may use random sampling to speed things up:
